Question 1 
Polynomial Regression

Calculate the sum square error for polynomial regression problem for
polynomial degree=2, using the follwing values of X and Y:
X: {0,1,2,3,1,1.5}
Y= {1,2,1.5,2,1,2}

```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
X=np.array([0,1,2,3,1,1.5])
Y=np.array([1,2,1.5,2,1,2])
```

```python
coe=np.polyfit(X,Y,2)
a,b,c=coe
print(f"alpha0:{a},  alpha1:{b},  alpha2:{c}")
Y_pred=a*X**2+b*X+c
rss = np.sum((Y_pred-Y)**2)
print(f"sum square error for polynomial: {rss}")
```

Question 2 Data Visualization

Generate a synthetic dataset by taking X and Y values as:
X: 6 random values in the range (1,3).
Y: generate the values using y= 1.7+ 0.5*x


a) Find out cost funcion value for 3 iterations.

```python
np.random.seed(42)
X = np.random.uniform(1, 3, size=(6,))
Y = 1.7 + 0.5 * X
learning_rate = 0.01
num_iterations = 3

m = 0.0
c = 0.0

def cost(y, y_pred):
    return np.mean((y - y_pred) ** 2)

def plot(X,Y,Y_pred):
    plt.plot(X, Y, color='black', label='Data Points', marker='D')
    plt.plot(X, Y_pred,color='red', label='Best fit',marker= 'D')
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.legend()
    plt.show()
    
for iteration in range(num_iterations):
    Y_pred = m * X + c
    Cost = cost(Y, Y_pred)
    dm = (2/len(X)) * np.sum(X * (Y_pred- Y))
    dc = (2/len(X)) * np.sum(Y_pred- Y)
    m = m - learning_rate * dm
    c = c - learning_rate * dc
    print(f"Iteration {iteration + 1},  Cost: {Cost},  m: {m},  c: {c}")
    plot(X,Y,Y_pred)
    

print("\nFinal Slope (m):", m)
print("Final intercept (c):", c)
```
